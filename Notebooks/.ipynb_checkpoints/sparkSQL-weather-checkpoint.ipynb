```python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession
from pyspark import SparkConf
from pyspark.sql.functions import *
```

```python
conf = SparkConf()\
    .set("spark.sql.warehouse.dir", "hdfs://C:/Users/mesteves/OneDrive - BI4ALL/Documents/MiniDatabricks/DeltaLake/HDFS")\
    .set("spark.sql.catalogImplementation", "hive")\
    .setMaster("local[*]")\
    .setAppName("Hive Example")

spark = SparkSession.builder\
    .config(conf=conf)\
    .enableHiveSupport()\
    .getOrCreate()
```

```python
print("c:\z/asdfa\asdasdf/".replace("\\", "/"))
```

```python
data = spark.read.format('csv').\
    option('inferSchema','true').\
    option('header', 'true').\
    option('path','C:\Users\mesteves\OneDrive - BI4ALL\Documents\MiniDatabricks\Setup Files\Datasets/GlobalWeatherRepository.csv').\
    load()
```

```python
data.printSchema()
```

```python
data.show()
```

```python
df = data.select(\
    "temperature_celsius", to_date("last_updated"))\
    .sort("country", "last_updated")\
    .filter("country = 'Afghanistan'"
)
df.show()
```

```python
df.toPandas().plot(kind='line',figsize=(12,6))
```
